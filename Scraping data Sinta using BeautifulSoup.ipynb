Then we extract all data with the program below

```python
pages = int(input("Max Pages: ")) #input many page want extract
#make empty list
no = []
titles = []
links = []
publishers = []
pissns = []
eissns = []
areas = []
sintas = []
scopus = []
garudas = []
impacts = []
hinds = []
c5ys = []
cits = []
#make dicti to save the list in csv with pandas
dicti = {"No":no, "Title":titles, "Link":links, "Publisher":publishers, "P-ISSN":pissns, "E-ISSN":eissns, "Subject Area":areas, "Sinta":sintas, "Scopus":scopus, "Garuda":garudas, "Impact":impacts, "H-5 Index":hinds, "Citations 5yr":c5ys, "Citations":cits}
numb = 0
#use for loop to looping the page
for i in range(1,pages+1):
    url = "https://sinta.kemdikbud.go.id/journals?page={0}".format(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    elements = soup.find_all("div",class_="list-item row mt-3") #get the specific element from html website
    for element in elements:
        title = (element.find("div", class_="affil-name mb-3").text).strip() #extract the title
        link = element.find("a").get("href") #extract the link
        publisher = (element.find("div",class_="affil-loc mt-2").text).strip() #extract publisher or author
        pissn = ((element.find("div",class_="profile-id").text).strip().replace(" ",'').split("\n")[0]).replace("|",'').split(":")[1] #extract the p-issn
        eissn = ((element.find("div",class_="profile-id").text).strip().replace(" ",'').split("\n")[1]).replace("|",'').split(":")[1] #extract the e-issn
        try: #using try when the element is found
            area = ((element.find("div",class_="profile-id").text).strip().replace(" ",'').split("\n")[2]).split(":")[1] #extract subject area
        except: #using except when the element is not found and then subject area is NaN
            area = ''
        indexs = (((element.find("div", class_="stat-prev mt-2").text).strip().replace(" ","")).replace("Accredited","")).replace("Indexed","").split("\n") #extract the index to make extracting sinta,scopus and garuda easier
        sinta = indexs[0] #extract the sinta
        if 'Scopus' in indexs: #extract the scopus when the scopus is found
            scop = "Yes"
        if 'Garuda' in indexs: #extract the garuda when the garuda is found
            garuda = "Yes"
        else: #extract scopus and garuda when not found
            scop = "No"
            garuda = "No"
        impact = ((((((element.find("div",class_="row no-gutters").text).strip().replace("\n","")).replace("Impact",":")).replace("H5-index",":")).replace("Citations 5yr",":")).replace("Citations",'')).replace(",",'.').split(":")[0] #extract the impact only numeric
        hind = (((((((element.find("div",class_="row no-gutters").text).strip().replace("\n","")).replace("Impact",":")).replace("H5-index",":")).replace("Citations 5yr",":")).replace("Citations",'')).replace(",",'')).replace(".",'').split(":")[1] #extract the h index only numeric
        c5y = (((((((element.find("div",class_="row no-gutters").text).strip().replace("\n","")).replace("Impact",":")).replace("H5-index",":")).replace("Citations 5yr",":")).replace("Citations",'')).replace(",",'')).replace(".",'').split(":")[2] #extract the citations 5yr only numeric
        cit = (((((((element.find("div",class_="row no-gutters").text).strip().replace("\n","")).replace("Impact",":")).replace("H5-index",":")).replace("Citations 5yr",":")).replace("Citations",'')).replace(",",'')).replace(".",'').split(":")[3] #extract the citations only numeric
        numb+=1
        no.append(numb) #save numb to no
        # save the extracted data to list
        titles.append(title) #save title to list titles
        links.append(link) #save link to list links
        publishers.append(publisher) #save publisher to publishers
        pissns.append(pissn) #save pissn to pissns
        eissns.append(eissn) #save eissn to eissn
        areas.append(area) #save subject area to areas
        sintas.append(sinta) #save sinta to sintas
        scopus.append(scop) #save scop to scopus
        garudas.append(garuda) #save garuda to garudas
        impacts.append(float(impact)) #save impact to impacts
        hinds.append(int(hind)) #save h index integer to h indexs
        c5ys.append(int(c5y)) #save citations 5yr integer to c5ys
        cits.append(int(cit)) #save citations integer to cits
    df = pd.DataFrame(dicti) #make dataframe with lists in above
    dfn = df.set_index("No") #set "No" as index
    dfn.to_csv("sinta's journal.csv") #save dataframe in csv format
    print("Page",i) #displays the page that has been extracted
print("End") #shows "End" if everything has been extracted
```

## Analyst Data 

```python
data = pd.read_csv("sinta's journal.csv") #read data from csv
data.info() #check data type
top_10_impact = data.sort_values(by="Impact",ascending=False).head(n=10) #sorting data by Impact
top_10_impact #show top 10 data after sort
```

```python
top_10_cite = data.sort_values(by="Citations",ascending=False).head(n=10) #sorting data by Citations
print(top_10_cite)
top_10_cite5yr = data.sort_values(by="Citations 5yr",ascending=False).head(n=10) #sorting data by Citations 5yr
print(top_10_cite5yr)
top_10_h5index = data.sort_values(by="H-5 Index",ascending=False).head(n=10) #sorting data by H-5 Index
print(top_10_h5index)
```
